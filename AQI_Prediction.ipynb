```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("city_hour.csv")
```

```python
df.tail(10)
```

```python
df.drop(['Datetime','AQI_Bucket','Benzene','Toluene','Xylene'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = 'NaN', strategy = 'mean')
imputer = SimpleImputer().fit(df.iloc[:, 1:14].values)
df.iloc[:,1:14] = imputer.transform(df.iloc[:, 1:14].values)
df.tail()
```

```python
df.drop_duplicates(subset=['City'],inplace = True)
```

```python
df.head(15)
```

```python
df.sort_values(by='City', ascending=True,inplace = True)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.City = le.fit_transform(dfle.City)
dfle.head(20)
```

```python
from sklearn.ensemble import AdaBoostRegressor 
```

```python
Train = df.drop(['AQI'],axis = 1)
Target = df['AQI']
```

```python
m2 = AdaBoostRegressor() 
m2.fit(Train, Target) 
```

```python
m2.score(Train, Target) * 100
```

```python
resultAQI = m2.predict([[1,67.622994,119.075804,1.00,40.01,36.370000,23.607959,1.00,122.070000,34.798979]])
```

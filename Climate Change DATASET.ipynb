```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
data = pd.read_csv("GlobalLandTemperaturesByCity.csv")
```

```python
data['AverageTemperature'].isnull().sum()
```

```python
data['AverageTemperatureUncertainty'].isnull().sum()
```

```python
data['AverageTemperature'] = data['AverageTemperature'].fillna((data['AverageTemperature'].mean()) )
```

```python
data['AverageTemperatureUncertainty'] = data['AverageTemperatureUncertainty'].fillna((data['AverageTemperatureUncertainty'].mean()) )
```

```python
dfIndia = data[data['Country'] == 'India']
```

```python
dfIndia.head()
```

```python
data['City'].isnull().sum()
```

```python
dfIndia.tail()
```

```python
start_yrs = '2000-01-01'
End_yrs = '2013-09-01'
```

```python
mask = (dfIndia['dt'] > start_yrs) & (dfIndia['dt'] <= End_yrs)
```

```python
dfIndia = dfIndia.loc[mask]
```

```python
dfIndia
```

```python
dfIndia.to_csv(r'C:\Users\AkshayK\Downloads\archive\ClimateChangeIndia.csv', index = False)
```

```python
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = dfIndia['AverageTemperature'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('AverageTemperature Change') 
ax.set_xlabel('Temperature') 
ax.set_ylabel('Frequency')
```

```python
dfIndia.sort_values("AverageTemperatureUncertainty",axis = 0,ascending = True,inplace = True, na_position = 'last')
```

```python
dfIndia.head(50)
```

```python
df = dfIndia[["dt","AverageTemperature","City"]]
```

```python
df.head()
```

```python
mask = (df['dt'] >= '2013-08-01') & (df['dt'] <= '2013-09-01')
```

```python
df = df.loc[mask] 
df.set_index("dt",inplace = True)
```

```python
df.describe()
```

```python
plt.figure(figsize = (10,7),dpi = 75)
plt.plot(df.index,df.AverageTemperature,color = 'tab:red')
plt.gca().set(title = "AVGTempChangeIndia 2008-2013",xlabel = "Date",ylabel = "Temp")
plt.show()
```
